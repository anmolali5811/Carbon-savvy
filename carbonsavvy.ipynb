## Perforoming Stratify Sampling on Data

```python
from sklearn.model_selection import StratifiedShuffleSplit
split = StratifiedShuffleSplit(n_splits=1,test_size=0.2,random_state=42)
for train_index,test_index in split.split(carbon_emission,carbon_emission['Garbage']):
    stratify_train_set = carbon_emission.loc[train_index]
    stratify_test_set = carbon_emission.loc[test_index]
```

```python
stratify_test_set['Garbage'].value_counts()
```

```python
stratify_train_set['Garbage'].value_counts()
```

## Hiding the Testing Data from the Models

```python
carbon_emission = stratify_train_set.copy()
carbon_emission.describe()
```

## Handling Missing Values

```python
co2_factors = stratify_train_set.drop("CO2_Emission",axis=1)
co2_factors = co2_factors.drop("Nuclear_Solar",axis=1)
co2_emission_labels = stratify_train_set["CO2_Emission"].copy()
co2_factors.head()
```

```python
co2_emission_labels.head()
```

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy="median")
imputer.fit(co2_factors)
co2_factors.describe()
```

```python
x = imputer.transform(co2_factors)
carbon_emission_tr = pd.DataFrame(x,columns = co2_factors.columns)
```

```python
carbon_emission_tr.describe()
```

## Creating Pipeline for new data

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
my_pipeline = Pipeline([
    ('imputer',SimpleImputer(strategy='median')),
    ('std_scaler',StandardScaler()),
])
carbon_emission_tr = my_pipeline.fit_transform(co2_factors)
```

## Selecting Model for the "CarbonSavvy"

### Training with Linear Regression

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(carbon_emission_tr,co2_emission_labels)
```

```python
some_data = co2_factors.iloc[:5]
some_labels = co2_emission_labels.iloc[:5]
prepared_data = my_pipeline.transform(some_data)
model.predict(prepared_data)
```

## Evaluating the Model

```python
from sklearn.metrics import mean_squared_error
import numpy as np
co2_predictions = model.predict(carbon_emission_tr)
```

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(model,carbon_emission_tr,co2_emission_labels,scoring="neg_mean_squared_error",cv=10)
rmse_scores = np.sqrt(-scores)
def print_scores(scores):
    print("Scores are: ",scores)
    print("Square root Mean error: ",scores.mean())
    print("std_deviation: ",scores.std())
print_scores(rmse_scores)
```

## Training with Decision Tree Regression

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error
import numpy as np
model = DecisionTreeRegressor()
model.fit(carbon_emission_tr,co2_emission_labels)
co2_emission_predictions = model.predict(carbon_emission_tr)
mse = mean_squared_error(co2_emission_labels,co2_emission_predictions)
rmse = np.sqrt(mse)
rmse
```

## Using Cross Validation Technique (KFold group data points)

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(model,carbon_emission_tr,co2_emission_labels,scoring="neg_mean_squared_error",cv=10)
rmse_scores = np.sqrt(-scores)
rmse_scores
```

```python
print_scores(rmse_scores)
```

## Training The model with Random forest Regressor

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import cross_val_score
import numpy as np
model = RandomForestRegressor()
model.fit(carbon_emission_tr,co2_emission_labels)
housing_predictions = model.predict(carbon_emission_tr)
scores = cross_val_score(model,carbon_emission_tr,co2_emission_labels,scoring="neg_mean_squared_error",cv=10)
rmse_scores = np.sqrt(-scores)
print_scores(rmse_scores)
```
